# Получение данных с сервера COCOAnnotator

```python
import json
import requests
from tqdm import tqdm 
```

```python
coco_admin_username = 'coco_admin'
coco_admin_password = 'AyhwhZsNEhurVfE8athc'

# Данные для авторизации в аннотатор
json_auth_data_admin = json.dumps({
    "password": coco_admin_password,
    "username": coco_admin_username
})

base_headers = {
    "Content-Type": "application/json",
    "accept": "application/json"
}

coco_base_URL = "http://130.193.54.188:5000/"

filtered_data = []
```

```python
def save_json_to_file(json_data, file_name):
    with open(file_name, 'w') as f:
        json.dump(json_data, f, indent=4)
```

```python
# Логинимся в аннотатор
s = requests.session()
login = s.post(coco_base_URL + "api/user/login",
               data=json_auth_data_admin,
               headers=base_headers)

if login.status_code == 200:
    raw_response = s.get(coco_base_URL + "/api/dataset", headers=base_headers)
    response = raw_response.json()

    for item in response:
        if item['name'].startswith("metric_book_"):
            filtered_data.append(item)
            
    print(f'Предварительная выборка метаданных для датасетов успешно завершена - загружено {len(filtered_data)} дата сета')
```

```python
# Получаем данные в формате COCO для каждого датасета
!mkdir /home/admin/notebooks/data_mb/images/datasets/coco
!mkdir /home/admin/notebooks/data_mb/images/datasets/coco/metadata_for_merge

for dataset in tqdm(filtered_data):
    coco_data = s.get(coco_base_URL + f"/api/dataset/{dataset['id']}/coco", headers=base_headers)
    save_json_to_file(coco_data.json(), f"/home/admin/notebooks/data_mb/images/datasets/coco/metadata_for_merge/{dataset['name']}.json")
    
print('COCO файлы в формате joson успешно скопированы и сохранены по пути - /home/admin/notebooks/data_mb/images/datasets/coco/metadata_for_merge')
```
